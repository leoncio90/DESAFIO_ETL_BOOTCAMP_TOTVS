```python
import json

def get_user(id):
    row = users[users['ID'] == id]
    return row.to_dict(orient='records')[0] if not row.empty else None

user = [user for id in user_ids if (user := get_user(id)) is not None]
print(json.dumps(user, indent=4, ensure_ascii=False))
```
